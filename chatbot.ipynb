```python
import re
import random
```

```python
class RuleBot:
  #potential negative response
  negative_response=("no","nope","nah","naw","not a chace","sorry")
  #exit commands
  Exit_command=("quit","pause","exit","goodbye","bye","later")
  #random question
  random_question=(
      "why are you here?",
      "Are there many humans like you?",
      "What do you consume life on this planet?",
      "Is there intelligent life on this planet?",
      "Does earth have a leader?",
      "What plaenets have you visited?",
      "what technology do you have on this planet?"
  )
  def __init__(self):
    self.alienbabble = {'describe_planet_intent':r'.*\s"your planet.*',
                    'answer_wht_intent': r'why\sare.*',
                      'about_intelipat': r'.*\s*intelipaat',
                        'about session': r'\s*session'}

  def greet(self):
    self.name=input("what is your name?\n")
    will_help = input(f"hii {self.name}, i am chat_bot .will you help me learn about planet?\n")
    if will_help in self.negative_response:
      print("ok,have a nice day!")
      return
    self.chat()

  def make_exit(self,reply):
    for command in self.Exit_command:
      if reply.strip() == command:
        print("ok,have a nice day!")
        return True
  def chat(self):
    reply=input(random.choice(self.random_question)).lower()
    while not self.make_exit(reply):
      reply=input(self.match_reply(reply))

  def match_reply(self, reply):
      for key, value in self.alienbabble.items():
          intent = key
          regex_pattern = value
          found_match = re.match(regex_pattern, reply)
          if found_match and intent == 'describe_planet_intent':
              return self.describe_planet_intent()
          elif found_match and intent == 'answer_wht_intent':
              return self.answer_wht_intent()
          elif found_match and intent == 'about_intelipat':
              return self.about_intelipat()
          elif found_match and intent == "about session":
              return self.about_session()
      if not found_match:
          return self.no_match_intent()


  def describe_planet_intent(self):
    responses=("my planet is a utopia of diverser origine and species.\n",
               "I am from opidipus,the capital of the wayward galxies.\n")
    return random.choice(responses)

  def answer_wht_intent(self):
    responses=("I come in peace.\n",
               "I am here to collect data on your planet and its inhabitants.\n",
               "I heard the coffe is good\n")
    return random.choice(responses)

  def about_intelipat(self):
    responses=("Intellipant is world's largest education company.\n",
               "Intellipaant will  make you learn concepts in the way you never learn.\n",
               "Intellipaant is where your carrer and skills grows\n")
    return random.choice(responses)

  def about_session(self):
    responses=("session is on 3 april.\n",
               "session was cool.\n")
    return random.choice(responses)

  def no_match_intent(self):
    responses=("Please tell me more.\n",
               "tell me more !.\n",
               "how do you think i feel whenyou say that ?\n")
    return random.choice(responses)
AlienBot =RuleBot()
AlienBot.greet()

```
